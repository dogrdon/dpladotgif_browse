```python
rawdatafile = './raw/dpladotgif_raw_dump.json'
```

```python
cleandatafile = './clean/dpladotgif_clean.json'
```

```python
data = json.loads(open(rawdatafile, 'r').read())
```

```python
cleandata = []
```

```python
data[:10]
```

```python
def list_join(l):
    if type(l) == list:
        return '--'.join(l)
    elif type(l) == str:
        return l
    else:
        return ''

def extract_nested(n):
    if n != '':
        return [i['name'].lower() for i in n]
    else: 
        return []


for d in data:
    clean_record = {'download_url' : d.get('record', {}).get('dl_url', ''),
                    'gif_loc' : d.get('gif', ''),
                    'score' : d.get('record', {}).get('score', ''),
                    'provider' : d.get('record', {}).get('provider_name', ''),
                    'provider_id' : d.get('record', {}).get('provider_id', ''),
                    'original_url' : d.get('record', {}).get('original_url', ''),
                    'dpla_id' : d.get('record', {}).get('dpla_id',''),
                    'search_term' : d.get('search_term',''),
                    'twitter_id' : d['twitter_post_id']['$numberLong'],
                    'video_duration' : d.get('video_duration', ''),
                    'title' : list_join(d.get('record', {}).get('source_resource', {}).get('title', '')),
                    'publisher' : d.get('record', {}).get('source_resource', {}).get('publisher', ''),
                    'creator' : list_join(d.get('record', {}).get('source_resource', {}).get('creator', '')),
                    'description' : d.get('record', {}).get('source_resource', {}).get('description',''),
                    'rights' : list_join(d.get('record', {}).get('source_resource', {}).get('rights','')),
                    'format' : list_join(d.get('record', {}).get('source_resource', {}).get('format','')),
                    'subjects' : extract_nested(d.get('record', {}).get('source_resource', {}).get('subject', ''))
                   }
    cleandata.append(clean_record)
    
```

```python
cleandata[:25]
```

```python
with open(cleandatafile, 'w') as outfile:
    json.dump(cleandata, outfile, indent=4)
```
